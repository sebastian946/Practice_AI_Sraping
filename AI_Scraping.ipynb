I just create a function call playwright and get the content that the AI gonna validate.

## Import section

```python
from scraping import run
from dotenv import load_dotenv
import os
import openai 
from openai import OpenAI
```

## Import .env variables

```python
load_dotenv(override=True)
api_key = os.getenv("AI_API_KEY")

if not api_key:
    raise ValueError("No API key found in .env file")
```

## Variables about the type of content (User, System, Model, Message)

```python


def message_validation(url: str):
    content = run(url)
    messages = [
        {"role": "system", "content": "You are a helpful assistant that validates the content of a website."},
        {"role": "user", "content": f"Please validate the content of the website this the content: {content} + the url: {url}"}
    ]
    return messages

model = "gpt-4.1-nano"
```

## Execute and complete the AI

```python
openai = OpenAI(api_key=api_key)
messages = message_validation("https://owasp.org/")
response = openai.chat.completions.create(model=model, messages=messages)

print(response.choices[0].message.content)
```
